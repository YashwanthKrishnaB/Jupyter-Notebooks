```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import BernoulliNB
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier,GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv('pima-indians-diabetes.csv')
data
```

```python
data.info()
```

```python
unique={col:data[col].nunique() for col in data.columns}
unique
```

```python
for col in data.columns:
    data[[col,'Outcome']].groupby('Outcome').mean().T.plot(kind='bar')
```

```python
sns.heatmap(data.corr(),annot=True)
```

```python
sns.distplot(data.Glucose)
```

```python
X=data.drop('Outcome',axis=1)
y=data['Outcome']
```

```python
train_x,test_x,train_y,test_y=train_test_split(X,y,test_size=0.3)
```

```python
accuracies=dict()
```

```python
models={
    LogisticRegression(),
    RandomForestClassifier(),
    AdaBoostClassifier(),
    GradientBoostingClassifier(),
    BernoulliNB(),
    SVC(),
    DecisionTreeClassifier()
}
```

```python
for model in models:
    model.fit(train_x,train_y)
    accuracies[model]=model.score(test_x,test_y)
accuracies
```

```python
model_acc={type(model).__name__:acc  for model,acc in accuracies.items()}
model_acc
```

```python
plt.bar(model_acc.keys(),model_acc.values())
plt.xticks(rotation=90);
```
