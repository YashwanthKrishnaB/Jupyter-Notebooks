```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', 500)
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('Customers_in_a_Shop.csv',header=None)
data.columns=['Date','Customers']
data['Date']=pd.to_datetime(data['Date'],format="%Y-%m")
data=data.set_index('Date')
data.head(140)
```

```python
data.plot()
```

```python
data.describe(include='all')
data.describe()
```

```python
data=data.ffill(axis=0)
data.head(140)
```

```python
data.describe(include='all')
```

```python
data.plot()
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
adfuller(data['Customers'],autolag='AIC')
```

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
m = 12
alpha = 1/(2*m)
```

```python
# airline[‘HWES1’] = SimpleExpSmoothing(airline[‘Thousands of Passengers’]).fit(smoothing_level=alpha,optimized=False,use_brute=True).fittedvalues
data['hwes1']=SimpleExpSmoothing(data['Customers']).fit(smoothing_level=alpha,optimized=False,use_brute=True).fittedvalues
data[['hwes1','Customers']].plot()
```

```python
# airline[‘HWES3_MUL’] = ExponentialSmoothing(airline[‘Thousands of Passengers’],trend=’mul’,seasonal=’mul’,seasonal_periods=12).fit().
data['hwes3-mul']=ExponentialSmoothing(data['Customers'],trend='mul',seasonal='mul',seasonal_periods=12).fit().fittedvalues
# data[['hwes3','114']].plot()
data[['Customers','hwes3-mul']].plot()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train=data[:120]
test=data[120:]
```

```python
model=ExponentialSmoothing(train['Customers'],trend='add',seasonal='add',seasonal_periods=12).fit()
predict=model.forecast(24)
train['Customers'].plot(legend=True,label='train')
test['Customers'].plot(legend=True,label='test')
predict.plot(legend=True,label='predicted')
```
