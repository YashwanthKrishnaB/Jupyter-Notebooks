```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

### Generating Random Input

n=1000
xs=np.random.uniform(low=-10,high=10,size=(n,1))
zs=np.random.uniform(-10,10,(n,1))
zs.shape

inputs=np.column_stack((xs,zs))
inputs.shape

### Generating ouput
# 2*x-3*z+5+noise

noise=np.random.uniform(-1,1,(n,1))
targets=2*xs-3*zs+5+noise
targets.shape



init_range=0.1

weights=np.random.uniform(-init_range,init_range,(2,1))
biases=np.random.uniform(-init_range,init_range,1)
print(weights,biases)



learning_rate=0.02

# Train the Model

for i in range(500):
    output=np.dot(inputs,weights)+biases
    deltas=output-targets
    loss=np.sum(deltas**2)/2/n
    print(loss)    
    deltas_scaled=deltas/n
    weights=weights-learning_rate*np.dot(inputs.T,deltas_scaled)
    biases=biases-learning_rate*np.sum(deltas_scaled)

print(weights,biases)

plt.plot(targets,output)
# plt.s
```
